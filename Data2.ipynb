```python
import statistics
def payment(arr):
  diff = arr[0]
  for i in range(1, len(arr)):
    diff += abs(arr[i] - arr[i-1])

  return diff


from itertools import permutations 
lists = list(permutations(range(1, 11))) 
output = []
for l in lists:
  perm = list(l)
  result = payment(perm)
  output.append(result)

print(len(output))

count = 0
for x in output:
  if x >= 45:
    count += 1
  

print(print("The probability that total payment is greater than or equal to 45 for N=20: % s "
        % (count/len(output))))

print("The mean of my total payments for N=10: % s "
        % (statistics.mean(output)))

print("The Standard Deviation of my total payments for N=10 % s " 
                % (statistics.stdev(output))) 

# The mean of my total payments for N=2: 2.5 
# The Standard Deviation of my total payments for N=2 0.7071067811865476

# The mean of my total payments for N=3: 4.666666666666667 // 2.1
# The Standard Deviation of my total payments for N=3 1.0327955589886444 // 0.4

# The mean of my total payments for N=4: 7.5 // 2.9
# The Standard Deviation of my total payments for N=4 1.5322617553657476 // 0.4

# The mean of my total payments for N=5: 11 // 3.5
# The Standard Deviation of my total payments for N=5 2.153753807236304 // 0.6

# The mean of my total payments for N=6: 15.166666666666666 // 4.1
# The Standard Deviation of my total payments for N=6 2.866525424817747 // 0.7

# The mean of my total payments for N=7: 20 // 4.9 
# The Standard Deviation of my total payments for N=7 3.651846020985047

# The mean of my total payments for N=8: 25.5 // 5.5
# The Standard Deviation of my total payments for N=8 4.500055804609463 // 0.9

# The mean of my total payments for N=9: 31.666666666666668 // 6.1
# The Standard Deviation of my total payments for N=9 5.405765695167962  // 0.9

# The mean of my total payments for N=10: 38.5 // 6.9
# The Standard Deviation of my total payments for N=10 6.365271230040156

# The mean of my total payments for N=11: 46 // 7.5
# The Standard Deviation of my total payments for N=11 7.375635658221922 

```
